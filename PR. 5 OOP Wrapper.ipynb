```python
class Person:
    # Constructor jo name aur age initialize karta hai
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Ye function person ke details print karega
    def show_details(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):
    # Constructor jo employee ke details initialize karega
    def __init__(self, name, age, employee_id, salary):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.salary = salary

    # Ye function employee ke details print karega
    def show_details(self):
        print(f"Name: {self.name}, Age: {self.age}, ID: {self.employee_id}, Salary: ${self.salary:.2f}")

class Manager(Employee):
    # Constructor jo manager ke details initialize karega
    def __init__(self, name, age, employee_id, salary, department):
        super().__init__(name, age, employee_id, salary)
        self.department = department

    # Ye function manager ke details print karega
    def show_details(self):
        print(f"Name: {self.name}, Age: {self.age}, ID: {self.employee_id}, Salary: ${self.salary:.2f}, Department: {self.department}")

# Salary compare karne ka function
def compare_salaries(emp1, emp2):
    if emp1.salary > emp2.salary:
        print(f"{emp1.name} ({emp1.employee_id}) ki salary {emp2.name} ({emp2.employee_id}) se zyada hai.")
    elif emp1.salary < emp2.salary:
        print(f"{emp2.name} ({emp2.employee_id}) ki salary {emp1.name} ({emp1.employee_id}) se zyada hai.")
    else:
        print(f"{emp1.name} ({emp1.employee_id}) aur {emp2.name} ({emp2.employee_id}) ki salary same hai.")

# Employee id same honi chahiye
def find_employee_by_id(employee_list, emp_id):
    for emp in employee_list:
        if emp.employee_id == emp_id:
            return emp
    return None

# Main function
def main():
    persons = []
    employees = []
    managers = []

    while True:
        print("\n--- Python OOP Project: Employee Management System ---")
        print("1. Create a Person")
        print("2. Create an Employee")
        print("3. Create a Manager")
        print("4. Show Details")
        print("5. Compare Salaries")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            persons.append(Person(name, age))

        elif choice == '2':
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            employee_id = input("Enter Employee ID: ")
            salary = float(input("Enter Salary: "))
            employees.append(Employee(name, age, employee_id, salary))

        elif choice == '3':
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            employee_id = input("Enter Employee ID: ")
            salary = float(input("Enter Salary: "))
            department = input("Enter Department: ")
            managers.append(Manager(name, age, employee_id, salary, department))

        elif choice == '4':
            print("Choose details to show:")
            print("1. Person")
            print("2. Employee")
            print("3. Manager")
            detail_choice = input("Enter your choice: ") # choice le raha h (1/2/3)

            if detail_choice == '1':
                for person in persons:
                    person.show_details() # person 1 ki detail show karega
            elif detail_choice == '2':
                for employee in employees:
                    employee.show_details() # Employee 2 ki detail show karega
            elif detail_choice == '3':
                for manager in managers:
                    manager.show_details()  # Manager 3 ki detail show karega
            else:
                print("Invalid choice.")

        elif choice == '5':
            if len(employees) + len(managers) < 2:
                print("Salary compare karne ke liye kam se kam 2 employees ya managers hone chahiye.")
            else:
                emp1_id = input("Enter the first employee's ID: ") # employee ki id le raha h
                emp2_id = input("Enter the second employee's ID: ")

                emp1 = find_employee_by_id(employees + managers, emp1_id) # ID match kr raha h
                emp2 = find_employee_by_id(employees + managers, emp2_id) # Id match kr raha h

                if emp1 and emp2:
                    compare_salaries(emp1, emp2)
                else:
                    print("Invalid employee IDs entered.") # id same nhi hui to invalid

        elif choice == '6':
            print("System exit ho raha hai. Goodbye!")
            break

        else:
            print("Invalid choice. Dubara try karein.")

if __name__ == "__main__": # starting point
    main()
```
