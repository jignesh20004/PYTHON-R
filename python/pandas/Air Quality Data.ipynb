line plot univariate plot

```python
plt.figure(figsize=(10, 5))
data[data['City'] == 'Delhi'].set_index('Datetime')['AQI'].plot(title='AQI Trend in Delhi')
plt.show()
```

Bar plot bivariate plot

```python
plt.figure(figsize=(10,5))
sns.barplot(x='City', y='AQI', data=data.head(10))
plt.title('Top ')
```

scatter plot bivariate 

```python
city_data = data[data['City'] == 'Delhi']  # Filter for a specific city

plt.figure(figsize=(10, 5))
plt.scatter(city_data['PM2.5'], city_data['PM10'], color='red', alpha=0.5)
plt.title('PM2.5 vs PM10 in Delhi')
plt.xlabel('PM2.5 (µg/m³)')
plt.ylabel('PM10 (µg/m³)')
plt.grid(True)
plt.show()
```

histogram plot univariate

```python
city_data = data[data['City'] == 'Delhi']
plt.figure(figsize=(10, 5))
plt.hist(data['AQI'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of AQI Levels')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

catplot bivariate

```python
plt.figure(figsize=(12, 6))
sns.catplot(x='City', y='AQI', data=data, kind='strip', height=6, aspect=2)

plt.title('AQI Levels in Different Cities')
plt.xticks(rotation=90)
plt.show()
```
