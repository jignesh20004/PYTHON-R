```python

student = []
dictionary = {}

while True:
  print("\nMenu option")
  print("1.add student")
  print("2.display all student")
  print("3.update student")
  print("4.delete student")
  print("5.display subject offered")
  print("6.Exit")

  choice = input("Enter your choice: ")
  if choice == "1":
    student_id = input("Enter student ID: ")
    name = input("Enter student name: ")
    age = input("Enter student age: ")
    grade = input("Enter student grade:")
    date_of_brith = input("Enter student date of birth: ")
    subject = input("Enter student subject: ")

    student.append({
        "student_id": student_id,
        "name": name,
        "age": age,
        "grade": grade,
        "date_of_brith": date_of_brith,
        "subject": subject
    })

    print("Student added successfully!")
  elif choice == "2":
    if not student:
      print("No student found")
    else:
      for student_data in student:
        print(f"name: {student_data['name']}, age: {student_data['age']}, grade: {student_data['grade']}, date_of_brith: {student_data['date_of_brith']}, subject: {student_data['subject']}")
  elif choice == "3":
    name = input("Enter student name: ")
    for student_data in student:
      if student_data["name"] == name:
        print("what do you want to update")
        print("1.name")
        print("2.age")
        print("3.grade")
        print("4.date_of_brith")
        print("5.subject")
        update_choice = input("Enter your choice:")
        if update_choice == "1":
          student_data["name"] = input("Enter new name: ")
        elif update_choice == "2":
          student_data["age"] = input("Enter new age: ")
        elif update_choice == "3":
          student_data["grade"] = input("Enter new grade: ")
        elif update_choice == "4":
          student_data["date_of_brith"] = input("Enter new date_of_brith: ")
        elif update_choice == "5":
          student_data["subject"] = input("Enter new subject: ")
        else:
          print("Invalid choice")
        print("Student updated successfully!")
        break
    else:
        print("Student not found")
  elif choice == "4":
    name = input("Enter student name: ")
    for i, student_data in enumerate(student):
      if student_data["name"] == name:
        del student[i]
        print("Student deleted successfully")
        break
    else:
      print("Student not found")
  elif choice == "5":
    for student_data in student:
      print(f"subject: {student_data['subject']}")
  elif choice == "6":
    break
  else:
    print("Invalid choice")
print("Thank you for using our system")
```
