```python
import math  # Math module import kar rahe hain taki factorial function use kar sakein

# Function jo user se data input lega aur ek list me store karega
def input_data():
    data = [int(x) for x in input().split()]
    print("Data has been stored successfully!")
    return data  # Data return kar rahe hain


def display_summary(data):
    print("\nData Summary:")  # Data summary heading print kar rahe hain
    print(f"- Total elements: {len(data)}")  # counting total element
    print(f"- Minimum value: {min(data)}")  # sab se small data print karega
    print(f"- Maximum value: {max(data)}")  # sab se bada data print karega
    print(f"- Sum of all values: {sum(data)}")  #  sab Data ka total
    print(f"- Average value: {sum(data)/len(data)}")

# Function jo recursion ka use karke factorial calculate karega
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)  # Recursive call: n ko (n-1)! se multiply kar rahe hain

# Function jo threshold ke basis par data ko filter karega (lambda function ka use karke)
def filter_by_threshold(data, threshold):
    filtered_data = list(filter(lambda x: x >= threshold, data))  # Threshold se chhoti values remove kar rahe hain
    print(f"Filtered Data (values >= {threshold}): {', '.join([str(x)for x in filtered_data])}")
    return filtered_data  # Filtered list return kr raha h

# Function jo user ke choice ke basis par data ko sort karega
def sort_data(data):
    sorting_option = input("Choose sorting option (asc/desc): ") # option asc small to big number / desc big to small number print
    sorted_data = sorted(data, reverse=True if sorting_option == "desc" else False)
    print("Sorted Data:", sorted_data)  # Sorted data print kar rahe hain

# Main function jo menu dikhayega aur respective functions call karega
def main():
    data = []  # Empty list initialize kar rahe hain data store karne ke liye
    while True:
        print("\nMain Menu:")
        print("1. Input Data")
        print("2. Display Data Summary (Built-in Functions)")
        print("3. Calculate Factorial (Recursion)")
        print("4. Filter Data by Threshold (Lambda Function)")
        print("5. Sort Data")
        print("6. Exit Program")
        choice = input("Please enter your choice: ")

        if choice == "1":
            data = input_data()  # Input function call kar rahe hain
        elif choice == "2":
            if data:
                display_summary(data)  # Data summary function call kar rahe hain
            else:
                print("No data available. Please input data first.")  # Agar data nahi hai to
        elif choice == "3":
            num = int(input("Enter a number to calculate its factorial: "))  # User se number input le rahe hain
            print(f"Factorial of {num} is: {factorial_recursive(num)}")  # Factorial calculate karke print kar rahe hain
        elif choice == "4":
            if data:
                threshold = int(input("Enter a threshold value to filter out data below this value: "))  # User se threshold input
                filter_by_threshold(data, threshold)  # Filter function call kar rahe hain
            else:
                print("No data available. Please input data first.")  # Agar data nahi hai to warning
        elif choice == "5":
            if data:
                sort_data(data)  # Sort function call kar rahe hain
            else:
                print("No data available. Please input data first.")  # Agar data nahi hai to warning
        elif choice == "6":
            print("Exiting program. Goodbye!")  # Exit message print kar rahe hain
            break  # Loop se bahar aa rahe hain aur program exit kar rahe hain
        else:
            print("Invalid choice. Please try again.")  # Agar user wrong input de to error message


# Program ka entry point
if __name__ == "__main__":  # Check kar rahe hain ki script direct run ho rahi hai ya nahi
    main()  # Main function call kar rahe hain


```
